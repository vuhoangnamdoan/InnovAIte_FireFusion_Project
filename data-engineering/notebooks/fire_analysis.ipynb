```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/fire_events.csv")

df['event_date'] = pd.to_datetime(df['event_date'])

print("Dataset Preview:")
df.head()

print("Dataset Info:")
df.info()

print("\nSummary Statistics:")
df.describe()

monthly = df.groupby(df['event_date'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
monthly.plot(kind='bar')

plt.title("Monthly Fire Incidents (Black Summer 2019–2020)")
plt.xlabel("Month")
plt.ylabel("Number of Fires")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

plt.figure(figsize=(10,6))

plt.scatter(df['longitude'], df['latitude'], s=1)

plt.title("Spatial Distribution of Fire Events (Australia)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()



df['confidence_score'].hist(bins=20)

plt.title("Confidence Score Distribution")
plt.xlabel("Confidence Score")
plt.ylabel("Frequency")

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()



```
